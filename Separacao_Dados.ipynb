## Tratamento inicial dos dados
O arquivo que constitui a base de dados original possui um tamanho de aproximadamente 330 Mb, por conta de limitações do GitHub, onde só é permitido um arquivo de até [100 Mb](https://docs.github.com/en/github/managing-large-files/what-is-my-disk-quota), logo se faz necessário dividir o arquivo original em vários para poder adicionar os dados ao repositório remoto.

```python
import pandas as pd
import numpy as np

df = pd.read_csv('data/deputies_dataset.csv')

df
```

Dividiremos o arquivo .csv original em um conjunto de arquivos onde cada um irá conter os dados de gastos para um determinado ano.

```python
df['year'] = pd.to_datetime(df['receipt_date']).dt.year # Adicionar uma coluna com o ano do gasto

df[['receipt_date', 'year']]
```

```python
for year in df['year'].unique(): # Para cada ano na base de dados
    df[df['year'] == year].to_csv('data/' + str(year) + '.csv') # Salvar todas as entradas daquele ano em um arquivo .csv
```
